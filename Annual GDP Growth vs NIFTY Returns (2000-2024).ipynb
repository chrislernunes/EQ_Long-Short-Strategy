```python
import matplotlib.pyplot as plt
import numpy as np

# Data
gdp_growth_data = {
    '2000': 3.84, '2001': 4.82, '2002': 3.80, '2003': 7.86, '2004': 7.92,
    '2005': 7.92, '2006': 8.06, '2007': 7.66, '2008': 3.09, '2009': 7.86,
    '2010': 8.50, '2011': 5.24, '2012': 5.46, '2013': 6.39, '2014': 7.41,
    '2015': 8.00, '2016': 8.26, '2017': 6.80, '2018': 6.45, '2019': 3.87,
    '2020': -5.78, '2021': 9.69, '2022': 6.99, '2023': 8.15, '2024': 8.20
}

nifty_returns_data = {
    '2000': -14.65, '2001': -16.18, '2002': 3.25, '2003': 71.90, '2004': 10.68,
    '2005': 36.34, '2006': 39.83, '2007': 54.77, '2008': -51.79, '2009': 75.76,
    '2010': 17.95, '2011': -24.62, '2012': 27.70, '2013': 6.76, '2014': 31.39,
    '2015': -4.06, '2016': 3.01, '2017': 28.65, '2018': 3.15, '2019': 12.02,
    '2020': 14.90, '2021': 24.12, '2022': 4.32, '2023': 19.42, '2024': 8.75
}

# Extract years and values
years = list(gdp_growth_data.keys())
gdp_growth = list(gdp_growth_data.values())
nifty_returns = list(nifty_returns_data.values())
```

```python
# Plotting
x = np.arange(len(years)) 
width = 0.35  

fig, ax = plt.subplots(figsize=(16, 8))

# Plot GDP growth bars
rects1 = ax.bar(x - width/2, gdp_growth, width, label='GDP Growth (%)', color='blue', alpha=0.8)

# Plot NIFTY returns bars
rects2 = ax.bar(x + width/2, nifty_returns, width, label='NIFTY Returns (%)', color='grey', alpha=0.8)

# Add labels, title, and legend
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Percentage (%)', fontsize=14)
ax.set_title('Annual GDP Growth vs NIFTY Returns (2000-2024)', fontsize=16, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(years, rotation=90, fontsize=10)
ax.legend(fontsize=12)

# Add a grid
ax.grid(True, linestyle='--', alpha=0.6, which='both', axis='y')  
ax.set_axisbelow(True) 


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

autolabel(rects1)
autolabel(rects2)


plt.tight_layout()
plt.show()
```
