```python
import yfinance as yf
import pandas as pd

nifty_ticker = "^NSEI" 

# Define the time period for the last 5 years
end_date = pd.to_datetime("today")
start_date = end_date - pd.DateOffset(years=5)
nifty_data_yf = yf.download(nifty_ticker, start=start_date, end=end_date)

nifty_data_yf['Returns'] = nifty_data_yf['Close'].pct_change()
nifty_data_yf['Day'] = nifty_data_yf.index.day
nifty_data_yf['Month'] = nifty_data_yf.index.month_name()

avg_returns_by_day_combined_yf = nifty_data_yf.groupby(['Day', 'Month'])['Returns'].mean().unstack()
avg_returns_by_day_combined_yf = avg_returns_by_day_combined_yf.reindex(index=range(1, 32), fill_value=0)
avg_returns_by_day_combined_yf = avg_returns_by_day_combined_yf.fillna(0)
avg_returns_by_day_combined_yf = avg_returns_by_day_combined_yf * 100

months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
avg_returns_by_day_combined_yf = avg_returns_by_day_combined_yf[months_order]

avg_returns_by_day_combined_yf = avg_returns_by_day_combined_yf.applymap(lambda x: f"{x:.2f}%")
avg_returns_by_day_combined_yf.reset_index(inplace=True)
avg_returns_by_day_combined_yf.columns.name = 'Day'
avg_returns_by_day_combined_yf.columns = ['Day'] + list(avg_returns_by_day_combined_yf.columns[1:])
avg_returns_by_day_combined_yf.head(31) 
```
